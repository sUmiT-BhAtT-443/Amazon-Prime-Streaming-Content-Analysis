Finally, the project explores **time-based patterns and engagement trends**, including decade-wise content growth and changes in genre distribution over time. These analyses help reveal how audience engagement, content quality, and popularity interact within the streaming ecosystem.

Overall, this exploratory analysis provides valuable insights into Amazon Prime’s content strategy, audience engagement patterns, and global content distribution. The findings offer a better understanding of how streaming platforms curate their libraries and how viewer preferences influence the evolution of digital entertainment.

### ***EDA(EXPLORATRYDATAANALYSIS)***

```python
#Replacing NaN values with 'Not Rated' in the age_certification column
title_df['age_certification'].fillna('Not Rated',inplace=True)

#Filling null values with mean values in imdb score column
title_df['imdb_score'].fillna(title_df['imdb_score'].mean(),inplace=True)


#Dropping duplicate values from both title_df and credit_df
title_df.drop_duplicates(inplace=True)
credit_df.drop_duplicates(inplace=True)

#Merging title_df dataset and credit_df dataset using 'id' as primary key
merged_df= title_df.merge(credit_df,on='id',how='left')

#Creating a column for decade
title_df['decade']=(title_df['release_year']//10)*10

#Categorize runtime into Short, Medium, or Long
merged_df['runtime_category']= ''
for i in range(len(merged_df)):
    if merged_df.loc[i, 'runtime']<30:
        merged_df.loc[i,'runtime_category']='Short'
    elif merged_df.loc[i,'runtime']<90:
        merged_df.loc[i,'runtime_category']='Medium'
    else:
        merged_df.loc[i,'runtime_category']='Long'

# Add column is_movie (1 if movie, 0 if show)
merged_df['is_movie']=1
for i in range(len(merged_df)):
    if merged_df.loc[i,'type']== 'MOVIE':
        merged_df.loc[i,'is_movie']=1
```

```python
#Categorize IMDb ratings (Poor, Average, Good, Excellent)

merged_df['rating_category']=''
for i in range(len(merged_df)):
    score= merged_df.loc[i,'imdb_score']
    if pd.isna(score):
        merged_df.loc[i,'rating_category']='Not Rated'
    elif score<5:
        merged_df.loc[i,'rating_category']='Poor'
    elif score<7:
        merged_df.loc[i,'rating_category']='Average'
    elif score<8.5:
        merged_df.loc[i,'rating_category']='Good'
    else:
        merged_df.loc[i,'rating_category']='Excellent'
```

### *GENERAL CONTENT OVERVIEW KPIS*

```python
#General Content Overview KPIs
total_titles=merged_df.shape[0]
total_movies=merged_df[merged_df['type']=='MOVIE'].shape[0]
total_shows=merged_df[merged_df['type']=='SHOW'].shape[0]
avg_runtime=merged_df['runtime'].mean()
avg_imdb=merged_df['imdb_score'].mean()
oldest_year=merged_df['release_year'].min()
newest_year=merged_df['release_year'].max()

print('TOTAL TITLES:',total_titles)
print('TOTAL MOVIES:',total_movies)
print('TOTAL SHOWS:',total_shows)
print('AVG RUNTIME:',round(avg_runtime,2))
print('AVG IMDB SCORE:',round(avg_imdb,2))
print('RELEASE YEARS RANGE:',oldest_year,'-',newest_year)
```

### ***GENRE BASED KPIs***

```python
# Split multiple genres into separate rows
genres_split=merged_df.dropna(subset=['genres']).copy()
genres_split['genres']=genres_split['genres'].str.replace('[','').str.replace(']','').str.replace("'","")
genres_split = genres_split.assign(genres=genres_split['genres'].str.split(','))
genres_split = genres_split.explode('genres')
genres_split['genres']=genres_split['genres'].str.strip()

print('TOP 10 GENRES...')
top_genres=print(genres_split['genres'].value_counts().head(10)) # Top 10 genres
print('----------------------------------------------------------------------------------------------------------')
print('AVG IMDB SCORE BY GENRE...')
avg_imdb_genre= genres_split.groupby('genres')['imdb_score'].mean().sort_values(ascending=False)   # Average IMDb by genre
print(avg_imdb_genre.head(10))
```
